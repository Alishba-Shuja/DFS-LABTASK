```python
# Graph with distances
graph = {
    'Arad': [('Zerind',75), ('Sibiu',140), ('Timisoara',118)],
    'Zerind': [('Oradea',71)],
    'Oradea': [('Sibiu',151)],
    'Sibiu': [('Fagaras',99), ('Rimnicu',80)],
    'Fagaras': [('Bucharest',211)],
    'Rimnicu': [('Pitesti',97)],
    'Pitesti': [('Bucharest',101)],
    'Timisoara': [('Lugoj',111)],
    'Lugoj': [('Mehadia',70)],
    'Mehadia': [('Drobeta',75)],
    'Drobeta': [('Craiova',120)],
    'Craiova': [('Pitesti',138)],
    'Bucharest': []
}

# DFS function
def dfs(node, goal, visited, path, cost):
    visited.add(node)
    path.append(node)

    if node == goal:
        print("Path:", " → ".join(path))
        print("Total Distance:", cost)
        return True

    for (neighbor, distance) in graph[node]:
        if neighbor not in visited:
            if dfs(neighbor, goal, visited, path, cost + distance):
                return True

    path.pop()
    return False

# Driver code
start = "Arad"
goal = "Bucharest"

dfs(start, goal, set(), [], 0)
```

🔹 Case 1: BFS visits fewer nodes than DFS
Initial State: Arad
Goal State: Bucharest

BFS explores level by level and reaches the goal faster through the shortest path.
DFS explores deep paths first and may visit unnecessary nodes before reaching the goal.

Therefore, BFS visits fewer nodes than DFS.

🔹 Case 2: DFS visits fewer nodes than BFS
Initial State: Lugoj
Goal State: Mehadia

DFS directly explores the deep path and reaches the goal quickly.
BFS explores all neighboring nodes level by level before reaching the goal.

Therefore, DFS visits fewer nodes than BFS.

```python
from collections import deque

# TSP Graph (simplified Romania map)
graph = {
    'Arad': ['Zerind', 'Sibiu', 'Timisoara'],
    'Zerind': ['Oradea'],
    'Oradea': ['Sibiu'],
    'Sibiu': ['Fagaras', 'Rimnicu'],
    'Fagaras': ['Bucharest'],
    'Rimnicu': ['Pitesti'],
    'Pitesti': ['Bucharest'],
    'Timisoara': ['Lugoj'],
    'Lugoj': ['Mehadia'],
    'Mehadia': ['Drobeta'],
    'Drobeta': ['Craiova'],
    'Craiova': ['Pitesti'],
    'Bucharest': []
}

# ---------------- BFS ----------------
def bfs(start, goal):
    visited = set()
    queue = deque([start])
    count = 0

    while queue:
        node = queue.popleft()

        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            count += 1

            if node == goal:
                print("\nBFS Nodes Visited:", count)
                return count

            for neighbor in graph[node]:
                queue.append(neighbor)

# ---------------- DFS ----------------
def dfs(node, goal, visited, count):
    visited.add(node)
    print(node, end=" ")
    count[0] += 1

    if node == goal:
        print("\nDFS Nodes Visited:", count[0])
        return True

    for neighbor in graph[node]:
        if neighbor not in visited:
            if dfs(neighbor, goal, visited, count):
                return True

    return False


# =========================
# CASE 1
# =========================
print("\n\nCASE 1: BFS < DFS")
print("BFS Traversal:")
bfs("Arad", "Bucharest")

print("\nDFS Traversal:")
dfs("Arad", "Bucharest", set(), [0])


# =========================
# CASE 2
# =========================
print("\n\nCASE 2: DFS < BFS")
print("BFS Traversal:")
bfs("Lugoj", "Mehadia")

print("\nDFS Traversal:")
dfs("Lugoj", "Mehadia", set(), [0])
```

```python
# DFS using STACK (Iterative)

graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

def dfs(graph, start):
    visited = set()
    stack = [start]

    print("Following is the Depth-First Search:")

    while stack:
        node = stack.pop()   # take top of stack

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            # Add neighbors to stack (reverse for correct order)
            for neighbour in reversed(graph[node]):
                if neighbour not in visited:
                    stack.append(neighbour)

# Driver code
dfs(graph, '5')
```

```python
from collections import defaultdict

# Graph class
class Graph:

    def __init__(self):
        self.graph = defaultdict(list)

    # Add edge
    def addEdge(self, u, v):
        self.graph[u].append(v)

    # DFS utility (recursive)
    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")

        for i in self.graph[v]:
            if not visited[i]:
                self.DFSUtil(i, visited)

    # Main DFS function
    def DFS(self):
        V = len(self.graph)
        visited = [False] * V

        for i in range(V):
            if not visited[i]:
                self.DFSUtil(i, visited)


# Driver code
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal")
g.DFS()
```
